```python
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
iris = datasets.load_iris()

df = pd.DataFrame(data = iris.data, columns = iris.feature_names)
df["target"] = iris.target

X = df.drop(["target"], axis = 1)
y = df["target"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .2, random_state = 42)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)

Dts = DecisionTreeClassifier(max_depth = 4)
Dts.fit(X_train, y_train)

y_train_pred = Dts.predict(X_train)
y_test_pred = Dts.predict(X_test)

y_train_accuracy = accuracy_score(y_train, y_train_pred)
y_train_confusion_matrix = confusion_matrix(y_train, y_train_pred)

y_test_accuracy = accuracy_score(y_test, y_test_pred)
y_test_confusion_matrix = confusion_matrix(y_test, y_test_pred)

print(y_train_accuracy)
print(y_test_accuracy)
```

```python
plt.figure(figsize = (7, 7))
sns.heatmap(data = y_train_confusion_matrix, 
            annot = True, 
            xticklabels = iris.target_names,
            yticklabels = iris.target_names,
            fmt = ".1f"
           )
plt.title("y_train Confusion Matrix")
plt.show()
```

```python
plt.figure(figsize = (7, 7))
sns.heatmap(data = y_test_confusion_matrix, 
            annot = True, 
            xticklabels = iris.target_names,
            yticklabels = iris.target_names,
            fmt = ".1f"
           )
plt.title("y_test Confusion Matrix")
plt.show()
```

```python
y_train_report = classification_report(y_train, y_train_pred, target_names = iris.target_names)
y_test_report = classification_report(y_test, y_test_pred, target_names = iris.target_names)
print(y_train_report)
print(y_test_report)
```
